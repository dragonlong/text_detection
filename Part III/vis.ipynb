```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
import os
```

```python
# design the path 
base_path = '/media/dragonx/DataStorage/ARC/EASTRNN/data/GAP_process'
video_list = []
for dir_name, dir_names, file_names in os.walk(base_path):
        for sub_dir_name in dir_names:
            video_list.append(sub_dir_name)
# load video and heatmap
# display the image sequentially 
# compare the effect 
print(video_list[0])
```

```python
video_path = base_path+'/'+video_list[0]+'.avi'
heatmap_path = base_path + '/'+video_list[0]+'/score'
frame_num = len(os.listdir(heatmap_path))
cap = cv2.VideoCapture(video_path)
j = 2
cap.set(j, 2)
ret, frame = cap.read()
print(frame.shape)
heatmap = np.squeeze(np.load(heatmap_path+'/frame'+'{0:03d}'.format(j)+'.npy'))
print(heatmap.shape)
```

```python
#visualization with cv2
%matplotlib inline
fig1 = plt.figure(figsize = (20,10))
#axes([0,0,0.7,0.6]) # add an axes with the position and size specified by 
                    # [left, bottom, width, height] in normalized units. 
fig1.add_subplot(1, 2, 1)
plt.imshow(frame)
fig1.add_subplot(1, 2, 2)
plt.imshow(heatmap)
plt.show()
print(heatmap[20,10])    
```
